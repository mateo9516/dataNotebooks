```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
print(iris.head())
```

1. Generar un grafico de dispersión con las variables sepal_length y sepal_width discriminando con la variable species

```python
sns.scatterplot(data = iris, x = "sepal_length", y = "sepal_width", hue = 'species', palette = 'colorblind')
```

2. Generar un gráfico de barras con la cantidad de species que hay

```python
sns.countplot(x = 'species', data = iris)
```

```python
df = pd.DataFrame(iris.species.value_counts()).reset_index()
sns.barplot(x = 'index', y= 'species', data =df)
```

```python
sns.histplot(iris, x = 'species')
```

3. Genere un diagrama de cajas y bigotes con species y petal_length

```python
sns.boxplot(x = 'species', y = 'petal_length', data = iris, palette = 'colorblind')
```

4. Genere un digrama de densidad (kdeplot) con species y la variable sepal_width

```python
sns.kdeplot(x = 'sepal_width', data = iris, hue = 'species')
```

5. (opcional) Genere un heatmap con la correlación de las variables numéricas.

```python
features = ['sepal_length','sepal_width','petal_length','petal_width']

corr = iris[features].corr()
sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns, cmap="Reds", annot=True, fmt=".2f");
```
