## Regresión Lineal

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

Empezaremos con la regresión lineal más conocida, un ajuste rectilíneo de los datos.
Un ajuste rectilíneo es un modelo de la forma
$$
y = ax + b
$$
donde $a$ se conoce comúnmente como la *pendiente*, y $b$ se conoce comúnmente como la *intercepto*.

Considera los siguientes datos, que están dispersos alrededor de una línea con una pendiente de 2 y un intercepto de -5:

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

Podemos utilizar el estimador ``LinearRegression`` de Scikit-Learn para ajustar estos datos y construir la línea de mejor ajuste:

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x.reshape(-1,1), y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

La pendiente y el intercepto de los datos están contenidos en los parámetros de ajuste del modelo, que en Scikit-Learn siempre están marcados con un guión bajo al final. Aquí los parámetros relevantes son coef_ e intercept_:

```python
print("Pendiente:    ", model.coef_[0])
print("Intercepto:", model.intercept_)
```

Sin embargo, el estimador ``LinearRegression`` es mucho más capaz que esto, ya que, además de los ajustes lineales simples, también puede manejar modelos lineales multidimensionales de la forma
$$
y = a_0 + a_1 x_1 + a_2 x_2 + \cdots
$$
donde hay múltiples valores de $x$.
Geométricamente, esto es similar al ajuste de un plano a puntos en tres dimensiones, o al ajuste de un hiperplano a puntos en dimensiones superiores.

La naturaleza multidimensional de estas regresiones hace que sean más difíciles de visualizar, pero podemos ver uno de estos ajustes en acción mediante la construcción de algunos datos de ejemplo, utilizando el operador de multiplicación de matrices de NumPy:

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])
```

```python
X
```

```python
y
```

```python
model.fit(X, y)
print(model.intercept_)
print(model.coef_)
```

## Ejemplo Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

The data contains the following columns:

* 'crim': per capita crime rate by town.
* 'zn': proportion of residential land zoned for lots over 25,000 sq.ft.
* 'indus': proportion of non-retail business acres per town.
* 'chas':Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
* 'nox': nitrogen oxides concentration (parts per 10 million).
* 'rm': average number of rooms per dwelling.
* 'age': proportion of owner-occupied units built prior to 1940.
* 'dis': weighted mean of distances to five Boston employment centres.
* 'rad': index of accessibility to radial highways.
* 'tax': full-value property-tax rate per $10,000.
* 'ptratio': pupil-teacher ratio by town
* 'black': 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
* 'lstat': lower status of the population (percent).
* 'medv': median value of owner-occupied homes in $$1000s

```python
BostonTrain = pd.read_csv("boston_train.csv")
```

```python
BostonTrain.head()
```

```python
BostonTrain.describe()
```

Qué variables son importantes?

```python
#ID no es importante
BostonTrain.drop('ID', axis = 1, inplace=True)
```

```python
#observamos un patron lineal con medv vs rm
BostonTrain.plot.scatter('rm', 'medv')
```

Ahora observemos la correlacion de todas las variables...

```python
plt.subplots(figsize=(12,8))
sns.heatmap(BostonTrain.corr())
```

Centrémonos en la última línea, donde y = medv:

Cuando los tonos de rojo/naranja: cuanto más rojo es el color en el eje X, más pequeño es el medv. Correlación negativa
Cuando los colores son claros: las variables en el eje x e y, no tienen ninguna relación. Correlación cero
Cuando hay tonos de gris/negro: cuanto más negro es el color en el eje X, más alto es el valor med. Correlación positiva

```python
#primero las negativas
sns.pairplot(BostonTrain, vars = ['lstat', 'ptratio', 'indus', 'tax', 'crim', 'nox', 'rad', 'age', 'medv'])
```

```python
#positivas
sns.pairplot(BostonTrain, vars = ['rm', 'zn', 'black', 'dis', 'chas','medv'])
```

```python
## Entrenemos!!
X = BostonTrain[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'black', 'lstat']]
y = BostonTrain['medv']
```

```python
#librerias (overfit y metrics)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
#obtengamos los coeficientes para mayor interpretabilidad
coefficients = pd.DataFrame(lm.coef_,X.columns)
coefficients.columns = ['coefficients']
coefficients.sort_values(by="coefficients",ascending=False)
```

por una unidad de aumento de nox, el valor de la casa disminuye 'nox'*1000 (correlación negativa) unidad de dinero.

para una unidad que aumenta rm, el valor de la casa aumenta 'rm'*1000 (correlación positiva) unidad monetaria.

*1000 porque el medv está en 1000 y esto se aplica a las otras variables/coeficientes.
